# Sample plot jupyter notebook 

#1. Import all necessary python modules

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as myplot
```

 #2. Define constants

```python
pi = np.pi
startx = -2
endx = 2
number_of_points=24000
```

#3. Calculate x values and y values arrays for plotting

```python
x=2*pi*np.linspace(startx,endx,number_of_points+1)
y1array=np.sin(x)
y2array=np.sin(2*x)
# print(x[0])
# print(x[24000])
# print(y1array[24000])
```

#4. Create plot of sin(x) - red, sin(2x) - green and sin(x) + sin (2x) - blue

```python
myplot.plot(x,y1array,'r');
myplot.plot(x,y2array,'g');
myplot.plot(x,y1array+y2array,'b');
myplot.show()
```
